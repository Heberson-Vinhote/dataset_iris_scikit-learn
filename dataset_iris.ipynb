```python
from sklearn.datasets import load_iris


iris_data = load_iris()


```

```python
iris_df = pd.DataFrame(data = iris_data['data'], columns = iris_data['feature_names'])
```

```python
iris_df.head()
```

```python
iris_df['Iris type'] = iris_data['target']
```

```python
iris_df.head()
```

```python
iris_df['Iris type'].value_counts()
```

```python
iris_df['Iris name'] = iris_df['Iris type'].apply(lambda x: 'setosa' if x == 0 else ('versicolor' if x == 1 else 'virginica'))
```

```python
def plot_violin(y2, i):
  plt.subplot(2, 2, i)
  sns.violinplot(x='Iris name', y= y2, data=iris_df)
```

```python
plt.figure(figsize=(17, 12))

i = 1

for meansurement in iris_df.columns[:-2]:
  plot_violin(meansurement, i)
  sns.despine(offset=10, trim=True)
  i += 1
```

A partir dos grficos de violino acima podemos notar uma alta densidade de comprimento e largura das espcies sentosa, especialmente para comprimento de spala, comprimento de ptala e largura de ptala. Tambm podemos observar que os valores mdios e o intervalo interquartil para as medidas das ptalas so facilmente distinguveis, embora os valores das espcies virginica sejam mais dispersos.
