# Treinando o modelo de ML

````python
from sklearn.model_selection import train_test_split
````

````python
iris_df.keys()
````

````python
X = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)' ]]
y = iris_df['Iris name']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
````

````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
````

````python
knn.fit(X_train, y_train)
````

# Avaliando o Modelo

````python
y_pred = knn.predict(X_test)
````

````python
y_pred
````

# Calculando a Acurácia

````python
print(f'a precisão do nosso modelo com k=3 é: {knn.score(X_test, y_test)}')
````

````python
def print_confusion_matrix(confusion_matrix, class_names, figsize=(9, 7), fontsize=14):
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d", cbar=False)
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label', fontsize=12)
    plt.xlabel('Predicted label', fontsize=12)
    plt.title('Confusion Matrix', fontsize=16)
````

````python
from sklearn.metrics import classification_report, confusion_matrix
````

````python
print_confusion_matrix(confusion_matrix(y_test, y_pred), ['sentosa', 'versicolor', 'virginica'])
````

````python
print(classification_report(y_test, y_pred))
````

De acordo com o relatório de classificação, podemos observar uma muito boa precisão, precisão e recall.

Precisão significa quantas previsões foram corretas em relação ao número total de previsões para uma classe. Precisão = TP/(TP + FP). Para as classes sentosa e versicolor, o KNN alcançou uma precisão perfeita. Já para a classe virginica, a precisão foi de 90%, o que significa que, das previsões atribuídas à classe virginica, 90% estavam corretas. Mais precisamente, o modelo previu 10 flores como virginica, das quais 9 foram previsões corretas (TP) e 1 foi incorreta (FP). Precisão_virginica = 9/(9+1) = 0,9.
Recall (ou sensibilidade) significa quantas previsões corretas foram feitas em relação ao número real de instâncias de uma classe. Recall = TP/(TP + FN). Para a classe versicolor, o recall foi de 94%, o que significa que o modelo previu corretamente 15 flores versicolor (TP), enquanto 1 delas foi erroneamente atribuída à classe virginica (FN). Recall_versicolor = 15/(15 + 1) = 0,94.
Nesse caso, como não temos um grande desequilíbrio em nossa amostra, a acurácia serve como uma boa medida de desempenho do modelo.
