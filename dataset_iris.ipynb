# Treinando o modelo de ML

````python
from sklearn.model_selection import train_test_split
````

````python
iris_df.keys()
````

````python
X = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)' ]]
y = iris_df['Iris name']
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
````

````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
````

````python
knn.fit(X_train, y_train)
````

# Avaliando o Modelo

````python
y_pred = knn.predict(X_test)
````

````python
y_pred
````

# Calculando a Acurcia

````python
print(f'a preciso do nosso modelo com k=3 : {knn.score(X_test, y_test)}')
````

````python
def print_confusion_matrix(confusion_matrix, class_names, figsize=(9, 7), fontsize=14):
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names
    )
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d", cbar=False)
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label', fontsize=12)
    plt.xlabel('Predicted label', fontsize=12)
    plt.title('Confusion Matrix', fontsize=16)
````

````python
from sklearn.metrics import classification_report, confusion_matrix
````

````python
print_confusion_matrix(confusion_matrix(y_test, y_pred), ['sentosa', 'versicolor', 'virginica'])
````

````python
print(classification_report(y_test, y_pred))
````

De acordo com o relatrio de classificao, podemos observar uma muito boa preciso, preciso e recall.

Preciso significa quantas previses foram corretas em relao ao nmero total de previses para uma classe. Preciso = TP/(TP + FP). Para as classes sentosa e versicolor, o KNN alcanou uma preciso perfeita. J para a classe virginica, a preciso foi de 90%, o que significa que, das previses atribudas  classe virginica, 90% estavam corretas. Mais precisamente, o modelo previu 10 flores como virginica, das quais 9 foram previses corretas (TP) e 1 foi incorreta (FP). Preciso_virginica = 9/(9+1) = 0,9.
Recall (ou sensibilidade) significa quantas previses corretas foram feitas em relao ao nmero real de instncias de uma classe. Recall = TP/(TP + FN). Para a classe versicolor, o recall foi de 94%, o que significa que o modelo previu corretamente 15 flores versicolor (TP), enquanto 1 delas foi erroneamente atribuda  classe virginica (FN). Recall_versicolor = 15/(15 + 1) = 0,94.
Nesse caso, como no temos um grande desequilbrio em nossa amostra, a acurcia serve como uma boa medida de desempenho do modelo.
